# Triple

```python
def check (a,b):

    # 교집합 개념을 사용하기 위해 리스트인  a, b를 set 자료형으로 변경한다.
    s1=set(a)
    s2=set(b)

    if s1 ==s1.intersection(s2) :
        return 1
    else : return 0
```

```python
dic3={}

from itertools import combinations

combi = []
for pair in list2 : 
    for item in list1 : 
        if( not (item in pair) ) :
            combi.append( (pair[0],pair[1],item) )

for c in combi : 
    if(check(combinations(c,2),list2)) :  #부분집합이 list2에 있는 애들만 택한다.    
        i = 0
        for b in baskets2 : 
            if( (c[0] in b)  and  (c[1] in b)  and  (c[2] in b) ) :
                i+=1
        dic3[c] = i
```

```python
list3 =[]
for key, val in dic3.items():
    if(val>=5):
        list3.append(key)
        print("key = {key}, value={value}".format(key=key,value=val))
```

```python
def dif_par(pair1, pair2) :
    if( pair1[0] in pair2 or pair1[1] in pair2 ) :
        return 0
    else :
        return 1
```
