```python
s=[]
for i in range(1,10) : 
    s.append(i/10)
```

```python
s
```

```python
def fx(s,r,b):
    return 1 - pow( (1- pow(s,r)) , b )
```

```python
import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
fx_list =[]
r=3
b=10
result =0
for i in s : 
    result = fx(i,r,b)
    fx_list.append(result)
plt.plot(s,fx_list)

#1/2일때 s값은??
def get_s(fx,r,b) : 
    return (1 - ((1-fx)**(1.0/b)) )**(1.0/r)
print("임계치 실측값 : " +str( get_s(1/2,r,b) ))
print("임계치 추정값 : " +str ( (1/b)**(1/r)))
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
example = "The most effective way to represent documents as sets, for the purpose of identifying lexically similar documents is to construct from the document the set of short strings that appear within it."
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(example)
result = []
for w in word_tokens:
    if w not in stop_words:
        result.append(w)
print(word_tokens)
print(result)
```
