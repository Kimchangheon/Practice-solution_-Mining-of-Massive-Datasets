```r
집합에 대한 자카드 유사도
```

```r
jaccard = function(a,b){
    intersect(a,b)
    union(a,b)
    length(intersect(a,b)) / length( union(a,b) )
}
```

```r
jaccard(a,b)
```

```r
jaccard(b,c)
```

```r
jaccard(a,c)
```

```r
bag1 = c(1,1,1,2)
bag2 = c(1,1,2,2,3)
bag3 = c(1,2,3,4)
```

백에대한 자카드 유사도
백a와 b에 대한 자카드 유사도를 구하는 경우 교집합에서는 a,b에서 한 원소가 반복되는 최소 횟수만큼 해당 원소를 센다. 그리고 합집합에서는 a와 c에서 해당 원소가 반복되는 횟수를 더한다.

```r
get_bag_intersection  = function(bag1, bag2){
    r =c()
    for(i in intersect(bag1,bag2)){
        m1 = sum(bag1 == i)
        m2 = sum(bag2 == i)
        if(m1< m2){
            r = c(r, rep(i, m1 ) )

        }else{
            r = c(r, rep(i, m2 ) )

        }

    }
    return(r)
}
```

```r
get_bag_intersection(bag1, bag2)
```

```r
jaccard_b = function(a,b){
    length(get_bag_intersection(a,b)) / length( c(a,b) )
}
```

```r
jaccard_b(bag1, bag2)
```

```r
jaccard_b(bag3, bag2)
```

```r
jaccard_b(bag3, bag1)
```

```r
set.seed(1)
size_list = sample(x=1:100,size=5000,replace=T)
```

```r
U = rep(1:100)
```

```r
jaccard_list =c()
for(m in size_list){
    S = sample(U, size = m, replace=F)
    T = sample(U, size = m, replace=F)
    jaccard_list = c(jaccard_list, jaccard(S,T) )
}
```

```r
plot(jaccard_list)
```

S와 T의 intersection이 존재할 학률 p = m/n이다. 
왜나하면 S가 m개의 원소를 가지고 있다고 가정해보자. 부분집합 T를 만드려고 하는데  원소중 S의 원소가 적어도 하나만 존재하면 된다. 따라서 T의 원소를 택할때  전체 n개중에서 S에 속한(m개의 원소)를 택하면 된다. 
따라서 m/n이다.
